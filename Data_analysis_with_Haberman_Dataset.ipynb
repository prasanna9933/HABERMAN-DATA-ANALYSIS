# Multivariate Analysis

# 2D SCATTER PLOT

```python
# plotting the scatter plot with two features 
# scateer plot to understand the relation between features 
cancer.plot(kind='scatter', x='age', y='nodes');
plt.title("Age vs Nodes")
plt.show()
```

# Observations
1. In this scatter plot, large amount of data available at node 0
2. And after node 30, less amount of data is available 

```python
sns.set_style("whitegrid");
sns.FacetGrid(cancer, hue="status", size=4) \
   .map(plt.scatter, "age", "nodes") \
   .add_legend();
plt.title("scatter plot with age vs nodes")
plt.show();
```

# Observations
1. In this plot survival status has different colour but still it is not separate 
2. Patients who are older than 50 and have axil nodes greater than 10 has more chance to die.
3. It is very much less likely to have patients with axil nodes more than 30.

# Univariate Analysis

# Histogram PDF CDF

```python
#1-D scatter plot of axil nodes
# as we know one dimension plot, do not give clear visualisation of data
# lets see after plotting scatter plot with one dimension

cancer_1 = cancer.loc[cancer["status"] == 1]; # cancer_1 dataset holds the features of survived patients 
cancer_2 = cancer.loc[cancer["status"] == 2]; # cancer_2 dataset holds the features of not survived patients 

plt.plot(cancer_1["nodes"], np.zeros_like(cancer_1['nodes']), 'o', label="1") #plotting scatter plot of survived patients with node features
plt.plot(cancer_2["nodes"], np.zeros_like(cancer_2['nodes']), 'o', label="2") #plotting scatter plot of not survived patients with node features
plt.title("scatter plot with nodes", size=18)
plt.legend()
plt.show()
```

# Observations
1. As we can see in this plot, nothing make sense in this plot 
2. Data is overlapping over each other 
3. After node 30, less amount of data 
4. Between 0 and 30 there is huge amount of data overlapping over each other

```python
# with the help of seaborn, plotting histogram with age features of both survived and not survived patients
sns.FacetGrid(cancer, hue="status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.title("Histogram with age", size=18)
plt.show();
```

# Observations
1. In this histogram of age features ,age between 40 and 45 more number of people has died ,and age below 40 are more chances to    survive 
2. with age features, again data is overlapping as usual 
3. it is hard to make sense with histogram of age features 

```python
# with the help of seaborn, plotting histogram with nodes features of both survived and not survived patients
sns.FacetGrid(cancer, hue="status", size=5) \
  .map(sns.distplot, "nodes") \
   .add_legend();
plt.title("Histogram with nodes",size=18)
plt.show();
```

# Observations
1. in thiis histogram of node features, as we can see patients having 0 nodes has more chances to survive

```python
# with the help of seaborn, plotting histogram with year features of both survived and not survived patients
sns.FacetGrid(cancer, hue="status", size=5) \
   .map(sns.distplot, "year") \
   .add_legend();
plt.title("histogram with operation year", size=18)
plt.show();
```

# Observations
1. In this histograam of operation year, mor nummber of patients died whose operation done operation year between 58 and 66

```python
# computing pdf and cdf 
counts, bin_edges = np.histogram(cancer_1['nodes'], bins=10, density = True)
# compute pdf                                  
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges);
# compute cdf
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf,label='PDF'); # plotting pdf 
plt.plot(bin_edges[1:], cdf,label='CDF') # plotting cdf


counts, bin_edges = np.histogram(cancer_1['nodes'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf); # plotting pdf 

plt.legend(loc="best")
plt.title("pdf and cdf of survived", size=18)
plt.show();
```

# Observations
1. In this plot, we have plotted pdf and cdf of survived patients with two bins of histogram

```python
counts, bin_edges = np.histogram(cancer_1['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))   # compute pdf
print(pdf);
print(bin_edges)

#compute CDF
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label="pdf") # plotting pdf  
plt.plot(bin_edges[1:], cdf, label="cdf") # plotting cdf


plt.title("pdf and cdf of survived", size=18)
plt.legend()
plt.show();
```

# Onservations
1. In this ploot we have plotted pdf and cdf of survived patients only

```python
# Plots the CDF of node for survived patients and died patients

#1(survived patients)
counts, bin_edges = np.histogram(cancer_1['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf, label="pdf")
plt.plot(bin_edges[1:], cdf,label="cdf")


# 2(deid patients)
counts, bin_edges = np.histogram(cancer_2['nodes'], bins=10, 
                                 density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)
plt.legend()
plt.title("pdf vs cdf of both survival status" ,size=18)
plt.show();
```

# Observations
1. In this plot we have plotted both pdf and cdf of died patient and survived patient
